Prime Numbers between 1-200

```


print("Prime numbers between 1-200 are:")

for num in range(1,201):
   # all prime numbers are greater than 1
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num,end=' ')
```
