#program to get presence of sublist into list

```
#L=[1,5,6,4,1,2,3,5]
L=[1,5,6,5,1,2,3,6]
l=[1,1,5]

cnt=0
k=0

for i in range(len(l)) :
	for j in range(k,len(L)) :
		if l[i]==L[j] :
	
			cnt=cnt+1
			k=j+1
			break 
		else :
			continue


if cnt==len(l):
 		print(L,"=>It's match")
else :
   print(L,"=>It's gone")

```

#[1, 5, 6, 5, 1, 2, 3, 6] =>It's *gone*
