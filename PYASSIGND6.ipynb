```
class BankAcc:
	def __init__(self,name,balance):
		self.name=name
		self.balance=balance

	def dis(self):
		print("\n**** Welcome ",self.name," *****")
		print("\nYour account is created with name:",self.name,"	 Current balance:",self.balance)

	def deposit(self):
		amount=int(input("\nEnter Amount to be deposited:"))
		Bal=self.balance+amount
		print("\n**** congratulations ",self.name," Your money is deposited *****")
		print("\ncurrent balance:",Bal)
		return Bal		

	def withdraw(self,Bal):
		amount=int(input("\nEnter Amount to be withdraw:"))
		print("\n**** congratulations ",self.name," Your money is withdraw *****")
		print("\nWithdraw Amount:",amount,"		current balance (before withdraw):",Bal)
		BA=Bal-amount;
		print("\n 		Current balance (after withdraw) :",BA)
		return BA



num=int(input("\n\nHow many Account you want to add:"))

for i in range(0,num):
	print("\n*********Enter Details of Account holder :",i+1,"******")

	name=input("\nEnter the name of Account holder:")
	balance=int(input("\nEnter current balance:"))

	B=BankAcc(name,balance)
	B.dis()

	B1=B.deposit()

	B.withdraw(B1)

```

```
import math
class CONE:

	def __init__(self,radius,height):
		self.radius=radius
		self.height=height


	def display(self):
		print("Radius :",self.radius,"Height:",self.height)

	def calvolume(self):
		volume=math.pi*(self.radius*self.radius)*(self.height/3)
		print("\nVolume of cone is:",volume)

	def calarea(self):
		a=math.pi*self.radius
		b=self.radius+math.sqrt((self.height*self.height)+(self.radius*self.radius))
		area=a*b
		print("\nSurface area of cone is:",area)



C=CONE(2,4)
C.display()
C.calvolume()
C.calarea()
```
