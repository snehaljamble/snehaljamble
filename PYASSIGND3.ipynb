```

Alt=0;

print("*******Welcome we are ready for landing*****\n")
print("****Enter the altitude for safe landing of plane****Minimum altitude should be less than 1000ft****\n")
Alt=int(input("Enter the current altitude to land the plane:"));

if(Alt<=1000):
	print("safe to land the plane")

elif(Alt>2000 and Alt<=5000):
	print("come down to 1000ft for safe landing of plane")

elif(Alt>5000):
	print("Move around and try later for safe landing of plane")

```
